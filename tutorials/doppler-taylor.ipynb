## Relativistic Doppler Effect - Taylor Expansion

The equation for the relativistic Doppler effect is below:

$$\lambda_{\rm obs}=\sqrt{\frac{1+v/c}{1-v/c}}\lambda_0$$

Taylor-expand this, assuming that $v/c$ is the small parameter.

Let the independent variable be $\beta = \frac{v}{c}$, and the dependent variable be $r = \frac{\lambda_{obs}}{\lambda_0}$.

Then, the equation becomes:

$$ r = \sqrt{\frac{1+\beta}{1-\beta}} $$

$$ r = (1+\beta)^{1/2}\cdot(1-\beta)^{-1/2} $$

For the first two terms of the Taylor expansion, find and evaluate the first two derivatives of $r(\beta)$ at $\beta=0$, since $\beta << 1$.

Evaluate $r$ at $0$:

$$ r(0) = 1 $$

Find the first derivative of $r(\beta)$:

$$ \frac{dr}{d\beta} = \frac{1}{2} (1+\beta)^{-1/2} \cdot (1-\beta)^{-1/2} + (-\frac{1}{2})(1-\beta)^{-3/2} \cdot (1+\beta)^{1/2} $$

$$ \frac{dr}{d\beta} = \frac{1}{2} (1+\beta)^{-1/2} \cdot (1-\beta)^{-3/2} \cdot ((1-\beta)-(1+\beta)) $$

$$ \frac{dr}{d\beta} = (1+\beta)^{-1/2} \cdot (1-\beta)^{-3/2} $$

Evaluate $\frac{dr}{d\beta}$ at $0$:

$$ \frac{dr}{d\beta}\bigg|_{\beta=0} = 1 $$

Find the second derivative of $r(\beta)$:

$$ \frac{d^2r}{d\beta^2} = (-\frac{1}{2})(1+\beta)^{-3/2}\cdot(1-\beta)^{-3/2} + (-\frac{3}{2})(1-\beta)^{-5/2}\cdot(1+\beta)^{-1/2} $$

$$ \frac{d^2r}{d\beta^2} = (-\frac{1}{2})(1+\beta)^{-3/2}\cdot(1-\beta)^{-5/2}\cdot((1-\beta) + 3(1-\beta)) $$

$$ \frac{d^2r}{d\beta^2} = (1+\beta)^{-3/2}\cdot(1-\beta)^{-5/2}\cdot(\beta-2) $$

Evaluate $\frac{d^2r}{d\beta^2}$ at $0$:

$$ \frac{d^2r}{d\beta^2}\bigg|_{\beta=0} = -2 $$

Then, the second-order Taylor expansion for $r = \sqrt{\frac{1+\beta}{1-\beta}}$ is:

$$ r \approx 1 + \frac{\frac{dr}{d\beta}(0)}{1!}(\beta-0) + \frac{\frac{d^2r}{d\beta^2}(0)}{2!}(\beta-0)^2 $$

$$ r \approx 1 + \beta - \beta^2 $$

Returning to the original variables,

$$ \frac{\lambda_{obs}}{\lambda_0} = 1 + \frac{v}{c} - (\frac{v}{c})^2 $$

Since $\frac{v}{c} << 1$, the $-(\frac{v}{c})^2$ term vanishes, leaving us with the classical Doppler shift equation.

$$ \frac{\lambda_{obs}}{\lambda_0} = 1 + \frac{v}{c} $$

Below, a comparison of the relativistic equation and the two Taylor expansions is shown:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot(upper_bound = 1, divs = 1000):
    beta = np.linspace(0, upper_bound, num=divs)
    r1 = 1+beta
    r2 = r1-beta**2
    r = np.sqrt((r1)/(1-beta))
    
    plt.figure(figsize=(10, 7))
    plt.ylabel('ratio of wavelengths')
    plt.xlabel('v/c')
    plt.plot(beta, r, label='relativistic')
    plt.plot(beta, r1, label='classical')
    plt.plot(beta, r2, label='2nd taylor')
    plt.legend()
```

```python
plot(0.1)
```

```python
plot(0.99)
```
