## Fourier Analysis

Proposed function:

$$ x(t)=\sum_{n=0}^{\infty}\frac{1-(-1)^{2n+1}}{(2n+1)^2}\cos((2n+1)t) $$

Below is the plot for the first $k$ functions in the series.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def series(k=5, lower=-2*np.pi, upper=2*np.pi, divs=10000):
    t = np.linspace(lower, upper, num=divs)
    x = np.zeros((k, divs))
    for n in range(k):
        a = 2*n+1
        x[n] = 2*np.cos(a*t)/a**2
    return t, x
```

```python
t, x = series(6)
for i in range(len(x)):
    plt.plot(t, x[i], label=str(i))

plt.title('Individual components')
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.show()
```

```python
for i in range(len(x)):
    if i > 0:
        x[i] += x[i-1]
    
    plt.plot(t, x[i], label=str(i))

plt.title('Sum of components')
plt.xlabel('t')
plt.ylabel('x')
plt.legend()
plt.show()
```

```python
plt.xlabel('t')
plt.ylabel('x')
plt.title('Final component')
plt.plot(t, x[len(x)-1])
plt.show()
```

This function seems to be approximating a triangle wave.
